```python
import numpy as np 
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

df = sns.load_dataset("tips")
df.head()

X = df[['total_bill']]
y = df['tip']

plt.title("Linear Regression", color="green")
plt.xlabel("total_bill", color="red")
plt.ylabel("tip", color="red")

# plt.scatter(X, y)
# plt.plot(df.total_bill, lr.predict(df[['total_bill']]), color="blue")

# plt.show()

sns.regplot(x=X, y=y, color="blue", line_kws={"color": "red"})

lr = LinearRegression()

lr.fit(df[['total_bill']], df.tip)

pred = lr.predict([[5000]])


print(f"predict value is: {pred}")
print(f"coefficient value is: {lr.coef_}")
print(f"intercept value is: {lr.intercept_}")

# from sklearn.model_selection import train_test_split

# X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)

# model = DecisionTreeRegressor()

# model.fit(X_train, y_train)

# y_pred = model.predict(X_test)

# from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error, r2_score

# print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred)}")
# print(f"Mean Absolute Percentage Error: {mean_absolute_percentage_error(y_test, y_pred)}")
# print(f"Mean Square Error: {mean_squared_error(y_test, y_pred)}")
# print(f"r2 score: {r2_score(y_test, y_pred)}")
# print(f"r2 score: {r2_score(y_test, y_pred)}")
```
